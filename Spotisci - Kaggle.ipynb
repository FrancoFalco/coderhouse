**Abstract**

The music industry underwent a profound transformation in the last decade, in harmony with the transition from the provision of goods to that of services that other sectors also experienced with the rise of the Internet and the digital revolution. This data science research project aims to analyze information from the well-known digital music _streaming_ server Spotify, a pioneering company that currently has a marked hegemony in the market, with the intention of generating an algorithmic model that allows predicting which songs They are more likely to be commercially successful after their release. For this purpose, it makes use of _Spotify popularity dataset_, a database belonging to the Kaggle site, which collects information on popularity scores prepared by Spotify for more than 40,000 songs, among other intrinsically musical variables. Under the hypothesis that the popularity of a certain song published on a _streaming_ service follows a predictable temporal evolution based on intrinsically musical characteristics, an exploratory data analysis was carried out that showed that the most popular songs preferentially presented: slight tendency towards the mode elderly; tempos between 80 and 140 BPM; Greater volume; indifference to tone; 4/4 time signature; under component of subjective variables "instrumental", "spoken", "live" and "acoustic"; high percentage of "danceability", "energy" and "valence". The selection of the most statistically significant variables during the _feature engineering_ ranked "Mode", "Volume", "Danceability (%)", "Spoken (%)", "Acoustic (%)" and "Instrumental (%)" as the most influential on the popularity score. For this reason, they were introduced into five machine learning classifier algorithms, namely: decision tree, logistic regression, random forest classifier, stacking algorithm (combining the first two) and neural network (deep learning model). _Random forest classifier_ emerged as the model with the best performance, with an _accuracy_ of 0.73, a precision of 0.73, a sensitivity of 0.90, an F1 _score_ of 0.80 and an area under the ROC curve of 0. .76. Although extra-musical factors could influence the popularity of a song to a significantly more robust extent, the present work demonstrates that certain intrinsically musical characteristics (such as volume or the perception of "danceability") play an important role. indisputable in the arrival of that popularity.

**Commercial context**

The music industry generated, globally, **$59.48 billion** in 2022, of which $15.9 billion was raised in the United States alone. The northern country is the most important market that currently supports this industry, followed by Japan, the United Kingdom and Germany.

At the same time, to date, only **9%** of music is sold in physical format. Even so, sales in physical format continued to represent around **19.2%** of the total income of the music industry in 2021. Likewise, sales through *streams* of subscribers and *streams* of advertising represented ** 47.3%** and **17.7%** of profits in the same period.

Between 2015 and 2016, an increase of approximately **64.7%** in subscriptions to streaming services was recorded, going from 68 million users to 112 million. Currently, the rate of increase in subscriptions is around **18%** annually and the number of registered users in 2022 is **616 million** worldwide.

This is a market characterized by a high concentration of capital on the supply side. **Universal Music Group** is currently the largest company globally, accounting for **32.1%** of total revenue in 2020. This figure even exceeds the sum of the profits of all independent labels , which in the same period corresponded to **31.4%**. **Sony Music Entertainment** and **Warner Music Group** are positioned behind, with **20.6%** and **15.9%** respectively.

On the demand side, **Spotify** is established as a dominant player, retaining around **31%** of the total subscribers to streaming services. Behind them are **Apple Music (15%)**, **Amazon Music (13%)**, **Tencent Music (13%)** and **YouTube Music (8%)**.

Despite all of the above, in 2017, 35% of internet users still accessed music through methods that involved copyright infringement, including stream-ripping (illegally saving music from streaming servers), and illegal downloads.

Finally, the most consumed genres both in physical format and on *streaming* platforms are **R&B/Hip Hop** with **28.2%** of sales in 2020, **Rock** with **19.5%** and **Pop** with **12.9%**. The remaining 39.4% is distributed among countless genres, among which country music, Latin music and electronic/dance music together account for almost 16% of sales [1].

**Business problem**

This work aims to identify, based on a historical and a transversal analysis of the characteristics of the music contained in streaming platforms, which are those that allow us to predict that a *track* will generate a high number of reproductions and/or interactions and, consequently, what attributes a song should have to be susceptible to such behavior.

It is worth clarifying that the fact of affirming that "the success" of a song is a direct consequence of its musical characteristics (such as its tempo, its compass, its mode or its tonality, among others) constitutes a reductionism that is difficult to find a correlation in the facts. , since, as is well known, countless other factors go into determining the popularity of a release. They are known as "socioeconomic factors." That is, a song with intrinsic conditions to become *mainstream* will hardly achieve it in the absence of other elements that contribute to musical dissemination, in which record labels invest a good part of their resources, such as like advertising. That is to say, the song is not an isolated event, but must be understood in the context of its release and in interaction with *extra-musical* factors.

However, a song that presents characteristics very distant from what is required to become *mainstream*, even in the presence of good publicity, could never achieve the expected success and even become a commercial failure. Hence the relevance of studying the preferences and medium-long term trends of the market, in order to guide all stages of musical production (composition, arrangement, recording, mixing, *mastering*, etc.) in the same direction.

**Analytical context**

In the quest to understand the dynamics behind song popularity and chart structure, different aspects have been explored over time. An early study, in 2000 [2], focused on the relationship between a song's position on the charts and its permanence on them, revealing that bigger hits tended to stay on for longer periods. Bhattacharjee et al. [3] developed a stochastic model that analyzed the longevity of albums on the *Billboard Top 100 Chart* and noted that, with the advent of file-sharing networks, the survival probability of songs on the charts decreased, but the Those who persisted experienced better prospects. Buda & Jarynowski [4], for their part, explored the trends of popular songs in 12 Western European countries, showing that geographical and cultural distances play different roles in the success of songs in different eras (analog, digital and Internet). ).

Many different factors have been studied in relation to the chart dynamics of songs, including popularity status ("superstar") and demographics of the artist, level of promotion, label/company, etc. These factors could be broadly classified as “socioeconomic” factors (as opposed to “musical” factors). Asai [5], for example, examined hits in Japanese popular music, highlighting artist fame, connections to other media, and the release of compilation albums as drivers of chart retention.

Other research has evaluated list dynamics from slightly different perspectives. Lee [6] analyzed the diversity of musical genres in the top 10 over the years, revealing a significant decrease in it between 1992 and 2002. Ordanini & Nunes [7] identified a decrease and subsequent increase in the number of songs on the *Billboard Hot 100*, attributing this phenomenon to changes in the music distribution industry with the emergence of legitimate online sellers.

In addition to socioeconomic factors, other authors have studied the influence of musical characteristics on the dynamics of the lists. Mauch et al. [8] considered the evolution of popular music using the US *Billboard Hot 100* charts between 1960 and 2010, and analyzed the musical properties of 17,000 recordings using data mining tools and measuring a series of quantitative audio characteristics. The analysis of musical diversity revealed that it has not decreased and also that the gradual evolution of musical styles has been marked by a series of “revolutions.” Askin & Mauskapf [9], for their part, examined the musical characteristics of the songs contained in the same list during a similar period (1958-2013). Their statistical analysis found that several of the songs' musical characteristics are positively correlated with success (measured as reaching the top 1 or top 10 on the charts, or staying on the charts for a certain amount of time). The authors then explored song cultural networks (defined by the patterns formed by the musical attributes of a set of releases) and found that those songs that are too similar to their neighbors have a harder time climbing the charts, unless a song is especially unique.

The present study is part of this last line of work that investigates musical characteristics and tries, based on them, to explain the popularity of songs.

**Hypothesis**

- The popularity of a certain song published on a _streaming_ server follows a relatively predictable temporal evolution based on certain intrinsically musical characteristics of the song.

```python
# Defining the function to display a univariate analysis table
def univariado_info(df):
    # Creating a dataframe with specific columns
    df_info = pd.DataFrame(columns=['Count', 'Type', 'Missing', 'Unique', 'Numeric'])
    
    # Looping through all dataframe variables
    for col in df.columns:  # Iterate over columns
        # Getting column information
        data_series = df[col]
        # Populating dataframe with initial columns
        df_info.loc[col] = [data_series.count(), data_series.dtype, data_series.isnull().sum(), data_series.nunique(), is_numeric_dtype(data_series)]

    # Calculating the describe
    numeric_columns = df.select_dtypes(include=['number']).columns
    if not numeric_columns.empty:
        df_describe = df[numeric_columns].describe().T[['mean', 'std', 'min', '25%', '50%', '75%', 'max']]
    else:
        df_describe = pd.DataFrame()

    # Calculating skewness and kurtosis (only for numeric columns)
    if not numeric_columns.empty:
        df_stats = pd.DataFrame([df[numeric_columns].skew(), df[numeric_columns].kurtosis()], index=['skewness', 'kurtosis']).T
    else:
        df_stats = pd.DataFrame()

    return pd.concat([df_info, pd.concat([df_describe, df_stats], axis=1)], axis=1).fillna('-')
```

```python
# Defining the function to identify Sturges intervals to be used in histograms
def sturges_intervals(data):
  # Sturges' rule for selecting the bin
  k = 1 + 3.322 * math.log10(len(data))
  periods = math.ceil(k)

  # Generating intervals based on range and periods
  intervals = pd.interval_range(
      start=data.min(),
      end=data.max() + 1,
      periods=k,
      name="Interval",
      closed="left")

  # Rounding the cut-off values
  rounded_intervals = pd.IntervalIndex([
    pd.Interval( round(i.left,1), round(i.right,1), i.closed )
    for i in intervals])

  return rounded_intervals, periods
```

```python
# Defining the function to create frequency tables
def freq_table(df, col, with_cuts=True):

  # Data selection
  data = df[col]
  #data =df[df[col]<130][col]
  # Checking column type and whether we want interval cuts
  if not is_numeric_dtype(data) or not with_cuts:
    # Generating frequency table with absolute info
    freq_tab = pd.crosstab(data, columns='FreqAbs').sort_values('FreqAbs', ascending=False)

  else:

    # Generating dataframe with cuts
    freq_tab = pd.DataFrame(index=sturges_intervals(data)[0])
    freq_tab["FreqAbs"] = pd.cut(data, bins=freq_tab.index).value_counts()
    freq_tab["Midpoint"]  = freq_tab.index.mid

    freq_tab["LowerLimit"] = freq_tab.index.left
    freq_tab["UpperLimit"] = freq_tab.index.right


  # Calculation of the rest of the frequency table
  freq_tab['FreqRel'] = freq_tab['FreqAbs'] / freq_tab['FreqAbs'].sum()
  freq_tab[['CumAbsFreq', 'RelFreq']] = freq_tab[['FreqAbs','FreqRel']].cumsum()
  return freq_tab
```

Before carrying out the treatment of _outliers_, it is necessary to know the type of distribution of each variable, in order to define the most appropriate treatment method. For this we will use the Kolmogorov-Smirnov test.

```python
# Define X
X = df_csv.drop(columns=['Popularidad'])

# Excluding the categorical columns from outliers treatment
X_outliers = df_csv.drop(columns=['Popularidad', 'Tonalidad', 'Modo', 'Compás'])
```

```python
# Initializing an empty dictionary to store the results
distribution_types = {}

# Iterating through each column in the DataFrame
for col in X_outliers.columns:
    # Performing the Kolmogorov-Smirnov test
    _, p_value = kstest(X[col].dropna(), 'norm')
    
    # Setting the threshold for the significance level
    alpha = 0.05
    
    # Checking if the p-value is less than the significance level
    if p_value < alpha:
        distribution_types[col] = 'NOT normal'   
    else:
        distribution_types[col] = 'Normal'

# Displaying the results
for col, distribution_type in distribution_types.items():
    print(f"The variable '{col}' has a {distribution_type} distribution.")
```

Given the type of **non-normal** distribution of the quantitative variables, it is preferred to use the interquartile range (IQR) as a parameter for detection and cleaning of _outliers_.

```python
# Defining a function to apply the IQR method to a single variable
def detectar_y_eliminar_atipicos_iqr(data):
    # Calculating the first and third quartile
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    
    # Calculating the interquartile range (IQR)
    IQR = Q3 - Q1
    
    # Defining the lower and upper limits for outliers
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    # Identifying outliers
    atipicos = (data < limite_inferior) | (data > limite_superior)
    
    # Counting the number of outliers
    num_atipicos = atipicos.sum()
    
    # Removing outliers from the data
    datos_limpios = data[~atipicos]
    
    return datos_limpios, num_atipicos


# Initializing an empty DataFrame to store the clean data
datos_limpios = pd.DataFrame()

# Iterating through each column in X
for col in X_outliers.columns:
    # Applying the IQR method to detect and remove outliers for the current column
    col_limpias, num_atipicos = detectar_y_eliminar_atipicos_iqr(df_csv[col])
    
    # Storing the clean column in the clean data DataFrame
    df_csv[col] = col_limpias

# Deleting rows with outlier values
df_csv = df_csv.dropna()

# Resetting the indexes
df_csv = df_csv.reset_index(drop=True)

# Printing the clean data
df_csv
```

**Stacking model**

We will combine the **decision tree** and **logistic regression** models in a ***stacking model***, with the intention of improving the general _performance_ of the algorithm, again using the _grid search_ technique to adjust the hyperparameters of the three models.

```python
# Individual models
tree = DecisionTreeClassifier(random_state=42)
logistic_model = LogisticRegression(random_state=42)

# Creating a Stacking Classifier using Random Forest as a meta-classifier
stacking_model = StackingClassifier(
    estimators=[('tree', tree), ('logistic', logistic_model)],
    final_estimator=RandomForestClassifier(random_state=42)
)

# Setting GridSearchCV for each model
tree_param_grid = {
    'max_depth': [5, 7, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

logistic_param_grid = {
    'C': [0.1, 1, 10],
    'max_iter': [50, 100, 200]
}

meta_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 15]
}

tree_grid_search = GridSearchCV(tree, tree_param_grid, cv=5)
logistic_grid_search = GridSearchCV(logistic_model, logistic_param_grid, cv=5)
meta_grid_search = GridSearchCV(stacking_model.final_estimator, meta_param_grid, cv=5)

# Fitting GridSearchCV on training data
tree_grid_search.fit(X_train, y_train)
logistic_grid_search.fit(X_train, y_train)
meta_grid_search.fit(X_train, y_train)

# Getting the best hyperparameters for each model
best_tree_params = tree_grid_search.best_params_
best_logistic_params = logistic_grid_search.best_params_
best_meta_params = meta_grid_search.best_params_

# Updating the models with the best hyperparameters
tree.set_params(**best_tree_params)
logistic_model.set_params(**best_logistic_params)
stacking_model.final_estimator.set_params(**best_meta_params)

# Tuning models with updated hyperparameters
tree.fit(X_train, y_train)
logistic_model.fit(X_train, y_train)
stacking_model.fit(X_train, y_train)

# Making predictions with the stacking model
stacking_preds = stacking_model.predict(X_test)

# Evaluating the performance of each individual model and the stacking model
print(f"Accuracy for Decision Tree Classifier: {accuracy_score(y_test, tree.predict(X_test)):.4f}")
print(f"Accuracy for Logistic Regression: {accuracy_score(y_test, logistic_model.predict(X_test)):.4f}")
print(f"Accuracy for Stacking Model: {accuracy_score(y_test, stacking_preds):.4f}")
```

```python
cv_scores = cross_val_score(stacking_model, X, y, cv=5, scoring='accuracy')

print("Cross-Validation Scores:", cv_scores)
print("Mean CV Score:", cv_scores.mean())
```
